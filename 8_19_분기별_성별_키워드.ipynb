```python
from konlpy.tag import Okt
from collections import Counter

# 형태소 분석기 초기화
okt = Okt()

# 제목에서 가장 많이 쓰인 단어 추출
titles = data_22_1_all['서명']
word_counts = Counter()

for title in titles:
    words = okt.nouns(title)  # 제목에서 명사 추출
    words = [word for word in words if len(word) > 1]  # 길이가 1인 단어 제외
    word_counts.update(words)


# 불필요한 단어 제거
unwanted_words = ['장편소설', '소설집', '소설','남매','카카오','프렌즈','설민석','쿠키런','어드벤처','코믹','메이플스토리','수학도둑']
for word in unwanted_words:
    if word in word_counts:
        del word_counts[word]

# 가장 많이 쓰인 단어 10개 추출
top_words = word_counts.most_common(20)

title_words_22_1_all = pd.DataFrame(top_words)
print(title_words_22_1_all)
```
